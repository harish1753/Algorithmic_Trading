```python
import yfinance as yf
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
# Load the data into a dataframe
symbol = yf.Ticker('BTC-USD')
df_btc = symbol.history(interval="1d",period="max")

# Filter the data by date
df_btc = df_btc[df_btc.index > datetime(2020,1,1)]
df_btc = df_btc[df_btc.index < datetime(2022,7,7)]

# Print the result
print(df_btc)
```

```python
del df_btc["Dividends"]
del df_btc["Stock Splits"]
```

```python
change = df_btc["Close"].diff()
change.dropna(inplace=True)
```

```python
# Create two copies of the Closing price Series
change_up = change.copy()
change_down = change.copy()

# 
change_up[change_up<0] = 0
change_down[change_down>0] = 0

# Verify that we did not make any mistakes
change.equals(change_up+change_down)

# Calculate the rolling average of average up and average down
avg_up = change_up.rolling(14).mean()
avg_down = change_down.rolling(14).mean().abs()
```

```python
rsi = 100 * avg_up / (avg_up + avg_down)

# Take a look at the 20 oldest datapoints
rsi.head(20)
```

```python
# Set the theme of our chart
plt.style.use('fivethirtyeight')

# Make our resulting figure much bigger
plt.rcParams['figure.figsize'] = (20, 20)
```

```python
# Create two charts on the same figure.
ax1 = plt.subplot2grid((10,1), (0,0), rowspan = 4, colspan = 1)
ax2 = plt.subplot2grid((10,1), (5,0), rowspan = 4, colspan = 1)

# First chart:
# Plot the closing price on the first chart
ax1.plot(df_btc['Close'], linewidth=2)
ax1.set_title('Bitcoin Close Price')

# Second chart
# Plot the RSI
ax2.set_title('Relative Strength Index')
ax2.plot(rsi, color='orange', linewidth=1)
# Add two horizontal lines, signalling the buy and sell ranges.
# Oversold
ax2.axhline(30, linestyle='--', linewidth=1.5, color='green')
# Overbought
ax2.axhline(70, linestyle='--', linewidth=1.5, color='red')
```

Combining all strategy 

```python
def MACD(Company):
    import numpy as np
    import yfinance
    from sklearn.preprocessing import StandardScaler
    import matplotlib.pyplot as plt
    company=yfinance.download(Company, start = "2010-01-01")
    company["EMA26"]=company.Close.ewm(span=26,min_periods=26).mean()
    company["EMA12"]=company.Close.ewm(span=12,min_periods=12).mean()
    company['MACD']=company['EMA12']-company['EMA26']
    company['Signal']=company.MACD.ewm(span=9,min_periods=9).mean()
    company['position(MACD)']=company['MACD']-company['Signal']
    company['position(MACD)']=np.where(company['position(MACD)']>0,1,0)
    company=company.dropna()
    return company['position(MACD)'].to_frame()
```

```python
def SMA(Company):
    import yfinance as yf
    import matplotlib.pyplot as plt
    from datetime import datetime
    import  numpy as np
    company = yf.download(Company, start = "2010-01-01")
    data=company.Close.loc[(company.index>='1991-01-01')]
    data=data.to_frame()
    sma_s=50
    sma_l=100
    data["sma_s"]=data.Close.rolling(sma_s).mean()
    data["sma_l"]=data.Close.rolling(sma_l).mean()
    data.dropna(inplace=True)
    data["position(SMA)"]=np.where(data["sma_s"]>data["sma_l"],1,0)
    data=data.dropna()
    return (data['position(SMA)']*0.3).to_frame()
```

```python
def EMA(Company):
    import yfinance as yf
    import matplotlib.pyplot as plt
    from datetime import datetime
    import  numpy as np
    company = yf.download(Company, start = "2010-01-01")
    data=company.Close.loc[(company.index>='1991-01-01')]
    data=data.to_frame()
    ema_s=50
    ema_l=100
    data["ema_s"]=data.Close.ewm(span=ema_s,min_periods=ema_s).mean()
    data["ema_l"]=data.Close.ewm(span=ema_l,min_periods=ema_l).mean()
    data.dropna(inplace=True)
    data["position(EMA)"]=np.where(data["ema_s"]>data["ema_l"],1,0)
    data=data.dropna()
    return (data['position(EMA)']*0.7).to_frame()
```

```python
def RSI(Company):
    import yfinance as yf
    import matplotlib.pyplot as plt
    from datetime import datetime
    import numpy as np
    import pandas as pd
    # Load the data into a dataframe
    symbol = yf.Ticker(Company)
    df = symbol.history(interval="1d",period="max")

    change = df["Close"].diff()
    change.dropna(inplace=True)
    # Create two copies of the Closing price Series
    change_up = change.copy()
    change_down = change.copy()

    # 
    change_up[change_up<0] = 0
    change_down[change_down>0] = 0

    # Verify that we did not make any mistakes
    change.equals(change_up+change_down)

    # Calculate the rolling average of average up and average down
    avg_up = change_up.rolling(14).mean()
    avg_down = change_down.rolling(14).mean().abs()
    rsi_data = 100 * avg_up / (avg_up + avg_down)
    rsi=rsi_data
    rsi=rsi.to_frame()
    rsi['0']=rsi_data*0
    rsi['1']=rsi_data/rsi_data
    rsi['0.5']=rsi_data/rsi_data*0.5
    rsi['position(RSI)']=rsi_data
    rsi['position(RSI)']=np.where(rsi_data>=30, rsi['0.5'],rsi['position(RSI)'])
    rsi['position(RSI)']=np.where(rsi_data>=70,rsi['1'],rsi['position(RSI)'])
    rsi['position(RSI)']=np.where(rsi_data>=80,rsi['0'],rsi['position(RSI)'])
    rsi['position(RSI)']=np.where(rsi_data<30, rsi['0'],rsi['position(RSI)'])
    rsi['position(RSI)']=np.where(rsi_data<=20, rsi['1'],rsi['position(RSI)'])
    rsi=rsi.dropna()
    return rsi['position(RSI)'].to_frame() 
```

```python
def accuracy(company):
    import yfinance as yf
    import matplotlib.pyplot as plt
    from datetime import datetime
    import numpy as np
    import pandas as pd
    
    symbol = yf.Ticker(company)
    df = symbol.history(interval="1d",period="max", start = "2010-01-01",end ="2021-01-01")
    change = df["Close"].diff()
    df['change']=change.dropna()
    print(df['change'])
    
    new_df=SMA(company).join(RSI(company))
    new_df=new_df.join(MACD(company))
    new_df=new_df.join(EMA(company))
    new_df=new_df.join(df['change'].to_frame())
    new_df['average']=new_df.mean(axis=1)*4/3
    new_df['decision2']=np.where(new_df['average']>=0.5,1,-1)
    new_df['decision1']=np.where(new_df['change']>=0,1,-1)
    new_df['decision_intermidiate']=new_df['decision1']*new_df['decision2']
    new_df['decision_final']=np.where(new_df['decision_intermidiate']<0,0,1)
    new_df=new_df.dropna()
    print(new_df)
    print('Precent accuracy of Model', company,':',(new_df['decision_final'].sum())/len(new_df['decision_final'])*100)
accuracy('INFY')
```

```python
def result(company):
    import numpy as np
    import matplotlib.pyplot as plt
    import matplotlib.dates as mdates
    new_df=SMA(company).join(RSI(company))
    new_df=new_df.join(MACD(company))
    new_df=new_df.join(EMA(company))
    new_df['average']=new_df.mean(axis=1)*4/3
    new_df['decision']=np.where(new_df['average']>=0.5,1,0)
    new_df.loc["2022-07",["position(SMA)","position(MACD)","position(RSI)",'position(EMA)',"decision"]].plot(figsize=(12,8),title=company, fontsize=12,secondary_y="decision")
    print(new_df.tail(25))
    return 'Sell' if new_df['decision'][-1]==0 else 'Buy/Hold'
```
